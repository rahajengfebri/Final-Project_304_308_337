# Cek Missing Value

```python
pd.isnull(df).sum()
```

Terdapat missing value pada kolom children, country, agent, dan company:  
1. Kolom `children` memiliki missing value terjadi karena tidak semua pelanggan yang memesan hotel membawa anak atau mungkin pelanggan tersebut tidak memiliki anak.  
2. Kolom `country` memiliki missing value terjadi karena sistem tidak mewajibkan pelanggan untuk mengisi informasi negara saat melakukan pemesanan, sehingga beberapa data tidak tercatat.  
3. Kolom `agent`memiliki missing value terjadi karena beberapa pemesanan dilakukan tanpa melalui agen perjalanan, seperti pemesanan langsung ke hotel melalui website atau telepon.  
4. Kolom `company` memiliki missing value terjadi karena beberapa pemesanan dilakukan oleh individu, bukan melalui perusahaan atau institusi yang bertanggung jawab atas pembayaran pemesanan.  

## Penanganan Missing Valaue

Berikut adalah penanganan missing value :  

| **Kolom**     | **Strategi Penanganan**                               |
|---------------|-------------------------------------------------------|
| `children`    | Imputasi atau mengganti nilai yang hilang dengan nilai 0 |
| `country`     | Imputasi atau mengganti nilai yang hilang dengan kategori "Unknown" |
| `agent`       | Imputasi atau mengganti nilai yang hilang dengan kategori "Direct Booking" |
| `company`     | karena hampir seluruh data pada kolom ini memiliki nilai yang hilang (missing value) mencapai sekitar 56.58% dari total dataset yang ada.<br> Selain itu, tidak ada kolom lain yang memiliki keterkaitan langsung dengan company, sehingga informasi yang ada pada kolom ini tidak <br> memberikan kontribusi signifikan terhadap hasil analisis  |

```python
import pandas as pd

# Menangani missing value pada kolom 'children'
df['children'] = df['children'].fillna(0)

# Menangani missing value pada kolom 'country'
df['country'] = df['country'].fillna('Unknown')

# Menangani missing value pada kolom 'agent'
df['agent'] = df['agent'].fillna('Direct Booking')

# Menangani missing value pada kolom 'company' dengan menghapus kolom
df.drop(columns=['company'], inplace=True)

```

## Problem Question 1.
Apakah terdapat pola dalam preferensi pelanggan terhadap tipe hotel tertentu? Hotel mana yang lebih sering dipilih oleh pelanggan, apakah Resort Hotel atau City Hotel?

```python
# Format visualisasi
sns.set(style="whitegrid", palette="muted", color_codes=True)

### 1. Analisis Tipe Hotel yang Sering Dipesan ###
hotel_counts = df['hotel'].value_counts()

# Visualisasi tipe hotel dengan angka di atas bar
plt.figure(figsize=(8, 5))
ax = sns.barplot(x=hotel_counts.index, y=hotel_counts.values, palette="viridis")
plt.title('Distribusi Pemesanan Berdasarkan Tipe Hotel')
plt.xlabel('Tipe Hotel')
plt.ylabel('Jumlah Pemesanan')

# Menambahkan angka di atas bar
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

- **Distribusi Pemesanan Berdasarkan Tipe Hotel**: Jumlah pemesanan untuk **City Hotel** mencapai **79.330**, sedangkan **Resort Hotel** hanya **40.060**. Perbedaan signifikan ini menunjukkan bahwa **City Hotel** lebih populer, kemungkinan karena lokasinya yang lebih strategis untuk perjalanan bisnis dan singkat, sementara **Resort Hotel** lebih jarang dipilih, yang umumnya dipilih untuk liburan yang lebih jarang dilakukan.

- **Pola Preferensi Pelanggan**: Pelanggan lebih sering memilih **City Hotel**, kemungkinan karena akses yang lebih mudah ke pusat bisnis. **Resort Hotel** lebih jarang dipilih, karena lebih cocok untuk liburan atau wisata, yang biasanya terjadi lebih jarang dibandingkan perjalanan bisnis.

- **Perbandingan Proporsi Pemesanan**: Sekitar **66,5%** dari total pemesanan memilih **City Hotel**, sementara **33,5%** memilih **Resort Hotel**. Proporsi ini menunjukkan dominasi **City Hotel** sebagai pilihan utama pelanggan, yang lebih sering dipilih untuk keperluan perjalanan bisnis daripada liburan.

- **Implikasi Potensial**: **City Hotel** memiliki peluang untuk terus menarik pelanggan yang fokus pada perjalanan bisnis, dengan memperkuat fasilitas yang mendukung kegiatan bisnis. Di sisi lain, **Resort Hotel** perlu mengevaluasi strategi pemasaran untuk meningkatkan pemesanan, terutama pada musim liburan atau akhir pekan, yang dapat meningkatkan kunjungan.

- **Kesimpulan**: **City Hotel** menjadi pilihan utama pelanggan berdasarkan data pemesanan, sedangkan **Resort Hotel** memiliki potensi untuk berkembang lebih jauh jika strategi pemasaran lebih efektif, terutama pada musim liburan.


## Problem Question 2.
Pada bulan dan tanggal berapa puncak pemesanan hotel biasanya terjadi? Apakah terdapat musim tertentu yang memengaruhi tingkat pemesanan hotel?

```python
### 2. Analisis Bulan dan Tanggal Puncak Pemesanan ###

# Membuat subplot 2 kolom
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Distribusi pemesanan per bulan
month_counts = df['arrival_date_month'].value_counts()

sns.barplot(x=month_counts.index, y=month_counts.values, order=month_counts.index, palette="mako", ax=axes[0])
axes[0].set_title('Distribusi Pemesanan Berdasarkan Bulan')
axes[0].set_xlabel('Bulan')
axes[0].set_ylabel('Jumlah Pemesanan')
axes[0].tick_params(axis='x', rotation=45)

# Menambahkan angka di atas bar
for p in axes[0].patches:
    axes[0].annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                     textcoords='offset points')

# Distribusi pemesanan per tanggal
day_counts = df['arrival_date_day_of_month'].value_counts()

sns.lineplot(x=day_counts.index, y=day_counts.values, marker="o", color="teal", ax=axes[1])
axes[1].set_title('Distribusi Pemesanan Berdasarkan Tanggal')
axes[1].set_xlabel('Tanggal')
axes[1].set_ylabel('Jumlah Pemesanan')

# Menambahkan angka di atas titik
for x, y in zip(day_counts.index, day_counts.values):
    axes[1].text(x, y + 1, f'{y}', ha='center', fontsize=10, color='black')

# Menambahkan spasi antara subplot
plt.tight_layout()
plt.show()
```

Pemesanan hotel di Eropa meningkat pada bulan Agustus, Juli, dan Mei:

2. **Peningkatan Permintaan Hotel**:

3. **Bulan Mei - Awal Musim Panas**:

| Langkah                             | Deskripsi                                                                                                                                    |
|-------------------------------------|------------------------------------------------------------------------------------------------------------------------------------------------|
| Menambah Kolom Baru             | Kolom baru 'season' ditambahkan dengan mengelompokkan bulan kedatangan tamu yang ada <br> di kolom 'arrival_date_month' ke dalam kategori musim. |
| Tujuan Penambahan Kolom         | Mempermudah analisis dengan mengelompokkan bulan-bulan dalam kategori musim, <br> sehingga tren musiman dapat terlihat dengan jelas tanpa memeriksa setiap bulan. |
| Musim Panas (Summer)            | Jika bulan kedatangan tamu adalah Juni, Juli, atau Agustus, <br> maka musim yang ditetapkan adalah "Summer".                                    |
| Musim Gugur (Fall)              | Jika bulan kedatangan tamu adalah September, Oktober, atau November, <br> maka musim yang ditetapkan adalah "Fall".                             |
| Musim Dingin (Winter)           | Jika bulan kedatangan tamu adalah Desember, Januari, atau Februari, <br> maka musim yang ditetapkan adalah "Winter".                            |
| Musim Semi (Spring)             | Jika bulan kedatangan tamu adalah Maret, April, atau Mei, <br> maka musim yang ditetapkan adalah "Spring".                                      |

Dengan mengetahui berapa banyak pemesanan yang terjadi di setiap musim, kita bisa mengidentifikasi kapan hotel cenderung lebih sibuk. Misalnya, jika musim panas atau liburan cenderung lebih banyak tamu yang datang, hotel bisa mempersiapkan diri dengan lebih baik. Ini juga membantu pihak manajemen hotel untuk membuat keputusan yang lebih tepat, seperti menawarkan promo di musim tertentu atau menambah staf untuk melayani lebih banyak tamu di waktu-waktu sibuk.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Fungsi untuk mengelompokkan bulan ke dalam musim
def categorize_season(month):
    if month in ['June', 'July', 'August']:
        return "Summer"
    elif month in ['September', 'October', 'November']:
        return "Fall"
    elif month in ['December', 'January', 'February']:
        return "Winter"
    elif month in ['March', 'April', 'May']:
        return "Spring"
    else:
        return "Unknown"

# Tambahkan kolom 'season' berdasarkan 'arrival_date_month'
df['season'] = df['arrival_date_month'].apply(categorize_season)

# Analisis pemesanan berdasarkan musim
booking_per_season = df.groupby('season')['is_canceled'].count().reset_index(name='total_bookings')

# Visualisasi jumlah pemesanan per musim
plt.figure(figsize=(10, 6))
sns.barplot(
    data=booking_per_season,
    x='season',
    y='total_bookings',
    palette='viridis'
)
plt.title("Analisis Jumlah Pemesanan Berdasarkan Musim", fontsize=14)
plt.xlabel("Musim", fontsize=12)
plt.ylabel("Jumlah Pemesanan", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Menampilkan data analisis musim
print("\nData jumlah pemesanan berdasarkan musim:")
print(booking_per_season)
```

Dari hasil visualisasi diatas, dapat dilihat bahwa pemesanan paling banyak terjadi pada musim summer, khususnya pada bulan Juni, Juli, dan Agustus. Ini menunjukkan bahwa banyak tamu yang memilih untuk menginap di hotel selama liburan musim panas, yang biasanya bertepatan dengan liburan sekolah atau waktu-waktu liburan yang lebih panjang.

Temuan Baru:<br>
Dalam analisis ini, kami membandingkan **Lead Time Category** berdasarkan **customer type**, yang meliputi kategori baru **transient-party**. Kategori **Lead Time** dibagi menjadi lima rentang waktu yang lebih detail, yaitu:

1. **0-7 Days**: Pemesanan yang dilakukan antara 0 hingga 7 hari sebelum kedatangan.
2. **8-30 Days**: Pemesanan yang dilakukan antara 8 hingga 30 hari sebelum kedatangan.
3. **31-100 Days**: Pemesanan yang dilakukan antara 31 hingga 100 hari sebelum kedatangan.
4. **101-365 Days**: Pemesanan yang dilakukan antara 101 hingga 365 hari sebelum kedatangan.
5. **More than 1 Year**: Pemesanan yang dilakukan lebih dari 365 hari sebelum kedatangan.

Perbandingan ini memungkinkan kita untuk melihat bagaimana jenis pelanggan (seperti **transient**, **contract**, **group**, dan **transient-party**) berperilaku dalam hal lead time mereka. Apakah mereka lebih cenderung melakukan pemesanan mendekati tanggal kedatangan (kategori **0-7 Days**) atau melakukan pemesanan lebih awal (kategori **More than 1 Year**). Dengan demikian, kita dapat memperoleh wawasan yang lebih lengkap mengenai pola pemesanan jangka pendek dan panjang yang bervariasi antar jenis pelanggan.

Dengan penambahan kategori yang lebih detail, analisis ini memberi gambaran yang lebih mendalam mengenai kecenderungan pemesanan pelanggan pada berbagai periode waktu.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Misalkan data sudah dimuat ke dalam DataFrame 'df'

# Membagi lead_time menjadi beberapa kategori dengan batas baru
bins = [0, 7, 30, 100, 365, float('inf')]  # Rentang kategori: 0-7, 8-30, 31-100, 101-365, >365
labels = ["0-7 Days", "8-30 Days", "31-100 Days", "101-365 Days", "More than 1 Year"]
df["Lead_Time_Category"] = pd.cut(df["lead_time"], bins=bins, labels=labels)

# Menghitung jumlah pemesanan untuk setiap kategori dan customer type
category_customer_counts = df.groupby(["customer_type", "Lead_Time_Category"]).size().unstack()

# Membuat bar chart
category_customer_counts.plot(kind='bar', stacked=False, figsize=(12, 8),
                              color=['#FF6F61', '#6B8E23', '#20B2AA', '#9370DB', '#FF6347'])

# Menambahkan judul dan label
plt.title("Comparison of Lead Time Categories by Customer Type", fontsize=14)
plt.xlabel("Customer Type", fontsize=12)
plt.ylabel("Number of Bookings", fontsize=12)

# Menambahkan legenda
plt.legend(title="Lead Time Category")

# Menampilkan grafik
plt.xticks(rotation=0)  # Agar label kategori customer type horizontal
plt.grid(True)
plt.show()
```

Berdasarkan grafik yang disajikan, hubungan antara **lead time** (jumlah hari antara tanggal pemesanan masuk ke sistem dan tanggal kedatangan) dengan **customer type** dapat dianalisis sebagai berikut:

1. **Customer Type "Contract"**
   - Jumlah pemesanan tercatat sangat kecil untuk semua kategori lead time.
   - Hal ini menunjukkan bahwa pemesanan tipe kontrak kemungkinan besar tidak bergantung pada lead time, karena biasanya dilakukan dalam skema perjanjian jangka panjang antara pihak hotel dan perusahaan atau organisasi tertentu.

2. **Customer Type "Group"**
   - Pemesanan tipe ini juga sangat sedikit di semua kategori lead time.
   - Fenomena ini mengindikasikan bahwa tipe grup tidak mendominasi dalam dataset yang dianalisis. Pemesanan grup cenderung dilakukan untuk keperluan acara tertentu yang memiliki pola lead time yang berbeda.

3. **Customer Type "Transient"**
   - Jenis pelanggan ini mendominasi jumlah pemesanan, terutama pada kategori lead time:
     - **31-100 Hari** dan **101-365 Hari**, dengan jumlah pemesanan tertinggi.
     - Pemesanan dalam kategori **0-7 Hari** juga signifikan, mengindikasikan adanya kebutuhan mendadak atau pemesanan last-minute.
   - Pola ini menunjukkan bahwa pelanggan transient cenderung memesan jauh hari, namun sebagian kecil juga memesan mendekati tanggal kedatangan.

4. **Customer Type "Transient-Party"**
   - Pola pada kategori ini mirip dengan "Transient" tetapi dengan jumlah yang lebih kecil.
   - Kategori lead time yang mendominasi tetap berada di rentang **31-100 Hari** dan **101-365 Hari**, dengan pemesanan dalam kategori lead time lebih pendek seperti **0-7 Hari** yang lebih jarang.

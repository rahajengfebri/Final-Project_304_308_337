## Pemesanan hotel berdasarkan musim


Untuk melakukan analisis berdasarkan musim, langkah pertama yang dilakukan adalah menambah kolom baru pada dataset, yang dinamakan kolom **'season'**. Kolom ini mengelompokkan bulan kedatangan tamu yang tercatat di kolom **'arrival_date_month'** ke dalam kategori musim yang sesuai. Tujuan dari penambahan kolom ini adalah untuk mempermudah analisis musiman, sehingga tren musiman dapat terlihat dengan jelas tanpa harus memeriksa setiap bulan secara terpisah.

Menambahkan kolom **`cancellation_status`** bertujuan untuk mengganti nilai numerik pada kolom **`is_canceled`** (0 untuk tidak dibatalkan dan 1 untuk dibatalkan) menjadi label yang lebih mudah dipahami, yaitu **'Dibatalkan'** dan **'Tidak Dibatalkan'**. Hal ini mempermudah analisis dan visualisasi status pembatalan pemesanan dengan cara yang lebih jelas dan mudah dimengerti.
Berikut adalah penjelasan mengenai pengelompokan musim berdasarkan bulan kedatangan tamu:

1. **Musim Panas (Summer)**: Jika bulan kedatangan tamu adalah **Juni, Juli, atau Agustus**, maka musim yang ditetapkan untuk tamu yang datang pada bulan-bulan tersebut adalah "Summer" atau Musim Panas. Musim ini biasanya identik dengan liburan sekolah dan waktu-waktu liburan panjang.

2. **Musim Gugur (Fall)**: Tamu yang datang pada bulan **September, Oktober, atau November** akan dikategorikan dalam musim "Fall" atau Musim Gugur. Musim ini biasanya ditandai dengan cuaca yang lebih sejuk, sering kali bertepatan dengan waktu anak-anak kembali ke sekolah.

3. **Musim Dingin (Winter)**: Jika tamu datang pada bulan **Desember, Januari, atau Februari**, maka musim yang ditetapkan adalah "Winter" atau Musim Dingin. Musim ini biasanya lebih dingin dan sering bertepatan dengan liburan akhir tahun atau musim liburan musim dingin.

4. **Musim Semi (Spring)**: Tamu yang datang pada bulan **Maret, April, atau Mei** akan dikelompokkan dalam musim "Spring" atau Musim Semi. Musim ini umumnya ditandai dengan cuaca yang lebih hangat, bunga-bunga mulai bermekaran, dan waktu liburan yang lebih tenang.

Dengan pengelompokkan ini, manajemen hotel dapat menganalisis tren musiman dengan lebih efektif dan mengambil keputusan berdasarkan pola permintaan dan pembatalan tingkat kunjungan pada setiap musim.

Dengan mengetahui berapa banyak pemesanan dan pembatalan pemesanan yang terjadi di setiap musim, kita bisa mengidentifikasi kapan hotel cenderung lebih sibuk. Misalnya, jika musim panas atau liburan cenderung lebih banyak tamu yang datang, hotel bisa mempersiapkan diri dengan lebih baik. Ini juga membantu pihak manajemen hotel untuk membuat keputusan yang lebih tepat, seperti menawarkan promo di musim tertentu atau menambah staf untuk melayani lebih banyak tamu di waktu-waktu sibuk.

**Cara Kerja**

1. Kolom `arrival_date_month` berisi nama bulan kedatangan tamu (misalnya, "January", "February", dll.).
   
2. Menambahkan Kolom Musim: Kolom baru `season` ditambahkan untuk mengelompokkan bulan kedatangan tamu ke dalam musim berdasarkan aturan berikut:
   - Summer (Panas): Juni, Juli, Agustus
   - Fall (Gugur): September, Oktober, November
   - Winter (Dingin): Desember, Januari, Februari
   - Spring (Semi): Maret, April, Mei
3. Hasil pengelompokan dipresentasikan menggunakan grafik batang untuk menunjukkan distribusi jumlah pemesanan dengan kategori permintaan pemesanan dan pembatalan pemesanan berdasarkan musim.


```python
# Fungsi untuk mengelompokkan bulan ke dalam musim
def categorize_season(month):
    if month in ['June', 'July', 'August']:
        return "Summer"
    elif month in ['September', 'October', 'November']:
        return "Fall"
    elif month in ['December', 'January', 'February']:
        return "Winter"
    elif month in ['March', 'April', 'May']:
        return "Spring"
    else:
        return "Unknown"

# Mengubah 'arrival_date_month' menjadi kolom 'season'
df['season'] = df['arrival_date_month'].apply(categorize_season)

# Menambahkan kolom 'cancellation_status' untuk status pemesanan dibatalkan atau tidak
df['cancellation_status'] = df['is_canceled'].apply(lambda x: 'Dibatalkan' if x == 1 else 'Tidak Dibatalkan')

# Menghitung jumlah pemesanan berdasarkan musim dan status pembatalan
booking_per_season = df.groupby(['season', 'cancellation_status']).size().reset_index(name='total_bookings')

# Visualisasi jumlah pemesanan per musim dengan status pembatalan
plt.figure(figsize=(10, 6))
sns.barplot(
    data=booking_per_season,
    x='season',
    y='total_bookings',
    hue='cancellation_status',  # Memisahkan batang berdasarkan status pembatalan
    dodge=True,  # Menampilkan batang terpisah untuk setiap kategori pembatalan
    palette={ 'Tidak Dibatalkan': '#20B2AA', 'Dibatalkan': '#FF6347'}  # Set warna untuk setiap kategori
)

plt.title("Analisis Jumlah Pemesanan Berdasarkan Musim dan Status Pembatalan", fontsize=14)
plt.xlabel("Musim", fontsize=12)
plt.ylabel("Jumlah Pemesanan", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Menambahkan legenda manual dengan warna yang sesuai
# Menentukan handle dan label sesuai dengan warna dan kategori
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='#20B2AA', markersize=10),
           plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='#FF6347', markersize=10)]
labels = ["Tidak Dibatalkan", "Dibatalkan"]

# Menambahkan legenda ke plot
plt.legend(handles=handles, labels=labels, title="Status Pembatalan", loc='upper right')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Menampilkan data analisis musim dan status pembatalan
print("\nData jumlah pemesanan berdasarkan musim dan status pembatalan:")
print(booking_per_season)
```

Dari hasil visualisasi diatas, dapat dilihat bahwa pemesanan paling banyak terjadi pada musim summer, khususnya pada bulan Juni, Juli, dan Agustus. Ini menunjukkan bahwa banyak tamu yang memilih untuk menginap di hotel selama liburan musim panas, yang biasanya bertepatan dengan liburan sekolah atau waktu-waktu liburan yang lebih panjang.

Pada musim summer, meskipun jumlah pemesanan tertinggi terjadi pada bulan Juni, Juli, dan Agustus, namun analisis menunjukkan bahwa pada musim ini juga terdapat tingkat pembatalan yang paling tinggi

## Pembatalan pemesanan city hotel dan resort hotel

Menambahkan kolom baru pada dataset, yang kita beri nama `cancellation_rate`. Kolom ini menunjukkan tingkat pembatalan sebagai persentase untuk setiap tipe hotel (Resort Hotel dan City Hotel).

Dengan menambahkan `kolom cancellation_rate`, dataset ini sekarang memiliki informasi tentang persentase pembatalan untuk setiap baris data, yang dihitung berdasarkan hotel masing-masing. Kolom ini membantu untuk menganalisis tingkat pembatalan berdasarkan jenis hotel, serta memberi gambaran lebih jelas tentang seberapa tinggi atau rendah tingkat pembatalan pada masing-masing kategori hotel.

Rumus untuk menghitung cancellation_rate berdasarkan variabel yang ada pada dataset adalah:
$$
\text{cancellation_rate} = \left( \frac{\text{df['is_canceled'].transform('mean')}}{\text{df.groupby('hotel')['is_canceled'].transform('mean')}} \right) \times 100
$$

Penjelasan Rumus :

- df['is_canceled'].transform('mean')
Fungsi transform('mean') pada df['is_canceled'] menghitung rata-rata nilai dari kolom is_canceled untuk seluruh dataset. Hasilnya adalah nilai rata-rata pembatalan (berapa persen pemesanan yang dibatalkan) di seluruh data.

- df.groupby('hotel')['is_canceled'].transform('mean')
Fungsi groupby('hotel') mengelompokkan data berdasarkan kolom hotel (misalnya, Resort Hotel atau City Hotel). Kemudian, transform('mean') dihitung untuk setiap grup hotel. Ini menghasilkan rata-rata pembatalan (berapa persen pemesanan yang dibatalkan) untuk setiap hotel secara terpisah.

- Membagi kedua nilai tersebut
Pembagian antara rata-rata pembatalan seluruh dataset dengan rata-rata pembatalan per hotel memberikan rasio yang menunjukkan seberapa tinggi tingkat pembatalan relatif untuk setiap hotel dibandingkan dengan keseluruhan dataset.

- Mengalikan dengan 100
Mengalikan hasil pembagian dengan 100 untuk mengonversinya ke dalam bentuk persentase.


**Cara Kerja**

1. Menambahkan Kolom `cancellation_rate`
  - Untuk setiap jenis hotel pada kolom `Hotel` (Resort Hotel dan City Hotel),  
    menghitung rata-rata pembatalan berdasarkan kolom is_canceled. Kolom is_canceled berisi angka 1 jika pemesanan dibatalkan dan angka 0 jika pemesanan tidak dibatalkan.
  - Dengan menggunakan metode groupby('hotel'), kita bisa menghitung rata-rata pembatalan untuk setiap hotel.
2. Menghitung Rata-rata Pembatalan untuk Setiap Hotel
  
  Memenghitung rata-rata pembatalan untuk setiap jenis hotel. Ini dilakukan dengan cara:
  Rata-rata Pembatalan per Hotel =
$$
\text{Rata-rata Pembatalan per Hotel} = \frac{\sum \text{Tingkat Pembatalan per Hotel}}{\text{Jumlah Pemesanan per Hotel}} \times 100
$$

  Penjelasan rumus:

  - Tingkat Pembatalan per Hotel dihitung berdasarkan rata-rata pembatalan untuk setiap pemesanan.
  - Jumlah Pemesanan per Hotel adalah jumlah pemesanan yang ada di setiap hotel.
  - Hasilnya dikalikan dengan 100 untuk mendapatkan persentase tingkat pembatalan per hotel.


3. Visualisasi Data
  Setelah kolom `cancellation_rate` ditambahkan, kita membuat grafik batang untuk memvisualisasikan perbandingan tingkat pembatalan antar jenis hotel.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Menambahkan kolom tingkat pembatalan
df['cancellation_rate'] = df.groupby('hotel')['is_canceled'].transform('mean') * 100

# Rata-rata tingkat pembatalan per tipe hotel
cancellation_rate = df.groupby('hotel')['cancellation_rate'].mean()

# Tentukan warna untuk setiap tipe hotel yang sesuai dengan palet
palette = sns.color_palette("coolwarm", 2)
color_map = {cancellation_rate.index[0]: palette[0], cancellation_rate.index[1]: palette[1]}

# Visualisasi
plt.figure(figsize=(8, 5))
ax = sns.barplot(x=cancellation_rate.index, y=cancellation_rate.values, hue=cancellation_rate.index, palette=color_map)
plt.title('Tingkat Pembatalan Berdasarkan Tipe Hotel')
plt.xlabel('Tipe Hotel')
plt.ylabel('Tingkat Pembatalan (%)')

# Menambahkan nilai persentase di atas bar
for index, value in enumerate(cancellation_rate.values):
    plt.text(index, value + 1, f'{value:.2f}%', ha='center')

# Menambahkan legenda manual dengan warna yang sesuai
handles = [
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color_map[cancellation_rate.index[0]], markersize=10),
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color_map[cancellation_rate.index[1]], markersize=10)
]
labels = ["City Hotel", "Resort Hotel"]

# Menambahkan legenda ke plot
plt.legend(handles=handles, labels=labels, title="Tipe Hotel", loc='upper right')

# Menampilkan grafik
plt.tight_layout()
plt.show()
```

- City Hotel menunjukkan tingkat pembatalan yang lebih tinggi. Ini bisa menjadi sinyal bagi manajemen hotel untuk mengevaluasi kebijakan pembatalan atau memahami lebih lanjut alasan di balik keputusan pembatalan yang tinggi.
- Resort Hotel memiliki tingkat pembatalan yang lebih rendah, yang bisa menandakan bahwa tamu di resort cenderung lebih konsisten dalam pemesanan mereka. Meskipun begitu, masih ada potensi pembatalan yang cukup besar, dan manajemen resort mungkin perlu mencari cara untuk mengurangi tingkat pembatalan lebih lanjut.

## Type Costumer berdasarkan lama Pemesanan

Dalam analisis ini, kami membandingkan **Lead Time Category** berdasarkan **customer type**, termasuk kategori baru **transient-party**. **Lead Time** dibagi menjadi lima rentang waktu yang lebih detail:  

1. **0-7 Days**: Pemesanan sangat mendekati tanggal kedatangan.  
2. **8-30 Days**: Pemesanan dalam waktu 1-4 minggu sebelum kedatangan.  
3. **31-100 Days**: Pemesanan dalam waktu 1-3 bulan sebelum kedatangan.  
4. **101-365 Days**: Pemesanan dalam waktu 3 bulan hingga 1 tahun sebelum kedatangan.  
5. **More than 1 Year**: Pemesanan lebih dari 1 tahun sebelumnya.  

Perbandingan ini memungkinkan untuk memahami pola pemesanan jangka pendek dan panjang dari berbagai jenis pelanggan (**transient**, **contract**, **group**, dan **transient-party**). Wawasan ini membantu dalam mengidentifikasi kebutuhan pelanggan pada berbagai periode waktu untuk menyusun strategi pemasaran atau pengelolaan sumber daya yang lebih efektif.

---

**Cara Kerja:**  
1. Kolom **`lead_time`** dibagi ke dalam kategori menggunakan fungsi **`pd.cut`**, dengan rentang kategori yang telah ditentukan. Kategori disimpan dalam kolom baru bernama **`Lead_Time_Category`**.  
2. Data kemudian dikelompokkan berdasarkan **`customer_type`** dan **`Lead_Time_Category`** menggunakan **`groupby`** dan dihitung jumlah pemesanan per kelompok.  
3. Hasil pengelompokan dipresentasikan menggunakan grafik batang untuk menunjukkan distribusi jumlah pemesanan per kategori lead time pada masing-masing jenis pelanggan.  
4. Grafik diberi label dan warna untuk memudahkan interpretasi, serta menampilkan bagaimana setiap jenis pelanggan berperilaku terhadap waktu pemesanan.

Hasil analisis ini memberikan wawasan penting bagi pengelola hotel dalam memahami preferensi pelanggan dan pola pemesanan mereka untuk meningkatkan pengalaman pelanggan dan optimasi bisnis.

```python
# Membagi lead_time menjadi beberapa kategori dengan batas baru
bins = [0, 7, 30, 100, 365, float('inf')]  # Rentang kategori: 0-7, 8-30, 31-100, 101-365, >365
labels = ["0-7 Days", "8-30 Days", "31-100 Days", "101-365 Days", "More than 1 Year"]
df["Lead_Time_Category"] = pd.cut(df["lead_time"], bins=bins, labels=labels)

# Menghitung jumlah pemesanan untuk setiap kategori dan customer type
category_customer_counts = df.groupby(["customer_type", "Lead_Time_Category"], observed=False).size().unstack()

# Membuat bar chart
category_customer_counts.plot(kind='bar', stacked=False, figsize=(12, 8),
                              color=['#FF6F61', '#6B8E23', '#20B2AA', '#9370DB', '#FF6347'])

# Menambahkan judul dan label
plt.title("Comparison of Lead Time Categories by Customer Type", fontsize=14)
plt.xlabel("Customer Type", fontsize=12)
plt.ylabel("Number of Bookings", fontsize=12)

# Menambahkan legenda
plt.legend(title="Lead Time Category")

# Menampilkan grafik
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

Berdasarkan grafik yang disajikan, hubungan antara **lead time** (jumlah hari antara tanggal pemesanan masuk ke sistem dan tanggal kedatangan) dengan **customer type** dapat dianalisis sebagai berikut:

1. **Customer Type "Contract"**
   - Jumlah pemesanan tercatat sangat kecil untuk semua kategori lead time.
   - Hal ini menunjukkan bahwa pemesanan tipe kontrak kemungkinan besar tidak bergantung pada lead time, karena biasanya dilakukan dalam skema perjanjian jangka panjang antara pihak hotel dan perusahaan atau organisasi tertentu.

2. **Customer Type "Group"**
   - Pemesanan tipe ini juga sangat sedikit di semua kategori lead time.
   - Fenomena ini mengindikasikan bahwa tipe grup tidak mendominasi dalam dataset yang dianalisis. Pemesanan grup cenderung dilakukan untuk keperluan acara tertentu yang memiliki pola lead time yang berbeda.

3. **Customer Type "Transient"**
   - Jenis pelanggan ini mendominasi jumlah pemesanan, terutama pada kategori lead time:
     - **31-100 Hari** dan **101-365 Hari**, dengan jumlah pemesanan tertinggi.
     - Pemesanan dalam kategori **0-7 Hari** juga signifikan, mengindikasikan adanya kebutuhan mendadak atau pemesanan last-minute.
   - Pola ini menunjukkan bahwa pelanggan transient cenderung memesan jauh hari, namun sebagian kecil juga memesan mendekati tanggal kedatangan.

4. **Customer Type "Transient-Party"**
   - Pola pada kategori ini mirip dengan "Transient" tetapi dengan jumlah yang lebih kecil.
   - Kategori lead time yang mendominasi tetap berada di rentang **31-100 Hari** dan **101-365 Hari**, dengan pemesanan dalam kategori lead time lebih pendek seperti **0-7 Hari** yang lebih jarang.

Kesimpulan
- Pelanggan **Transient** merupakan kelompok dominan dalam data ini, dengan lead time yang bervariasi tetapi cenderung memesan jauh hari (31-365 hari).
- Pelanggan dengan tipe **Contract** dan **Group** memiliki jumlah pemesanan yang jauh lebih kecil, sehingga tidak terlalu terpengaruh oleh faktor lead time.
- Lead time menjadi indikator penting dalam memprediksi kebutuhan kamar hotel, khususnya untuk pelanggan transient, sehingga dapat digunakan untuk menyusun strategi pemasaran yang lebih efektif.

## Lama menginap pelanggan pada City Hotel dan Resort Hotel

Dalam analisis ini, kami membandingkan rata-rata lama menginap tamu berdasarkan dua tipe hotel: City Hotel dan Resort Hotel. Lama menginap dihitung dengan menjumlahkan **stays_in_weekend_nights** dan **stays_in_week_nights**. Perbandingan ini memberikan wawasan mengenai kebiasaan tamu dalam memilih lama menginap, yang berguna untuk perencanaan pengelolaan sumber daya dan strategi pemasaran.

**Cara Kerja**

1. Menambahkan Kolom `total_stays`
   Menghitung total malam menginap dengan menjumlahkan **stays_in_weekend_nights** dan **stays_in_week_nights**.

2. Mengelompokkan Data dan Menghitung Rata-rata
   Data dikelompokkan berdasarkan **hotel** dan dihitung rata-rata **total_stays** untuk setiap tipe hotel.

3. *Visualisasi*
   - Membuat grafik batang (barplot) untuk membandingkan rata-rata lama menginap antara City Hotel dan Resort Hotel.

Hasil analisis ini memberikan gambaran tentang perbedaan kebiasaan lama menginap antara kedua tipe hotel, yang dapat digunakan untuk strategi pemasaran dan manajemen yang lebih efektif.

```python
import warnings

# Menyembunyikan semua warning
warnings.filterwarnings('ignore')

# Menambahkan variabel total lama menginap
df['total_stays'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']

# Distribusi total lama menginap per tipe hotel
total_stays_avg = df.groupby('hotel')['total_stays'].mean()

# Visualisasi
plt.figure(figsize=(8, 5))

# Menentukan warna untuk setiap tipe hotel
color_map = {'City Hotel': '#FF6347', 'Resort Hotel': '#20B2AA'}

# Membuat barplot untuk masing-masing tipe hotel dengan warna yang sesuai
sns.barplot(x=total_stays_avg.index, y=total_stays_avg.values, palette=color_map)

plt.title('Rata-rata Lama Menginap Berdasarkan Tipe Hotel')
plt.xlabel('Tipe Hotel')
plt.ylabel('Rata-rata Lama Menginap (malam)')

# Menambahkan legenda manual dengan warna yang sesuai
handles = [
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color_map['City Hotel'], markersize=10),
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color_map['Resort Hotel'], markersize=10)
]
labels = ["City Hotel", "Resort Hotel"]

# Menambahkan legenda ke plot
plt.legend(handles=handles, labels=labels, title="Tipe Hotel", loc='upper right')

plt.tight_layout()
plt.show()
```

- Berdasarkan hasil visualisasi, rata-rata lama menginap di **Resort Hotel** lebih lama dibandingkan dengan di **City Hotel**.

- Faktor Penyebab
  - Resort Hotel biasanya menawarkan fasilitas yang mendukung liburan panjang dan relaksasi, sehingga pelanggan cenderung menginap lebih lama.
  - City Hotel lebih sering digunakan oleh pelancong bisnis atau wisatawan yang hanya membutuhkan penginapan untuk waktu singkat, sehingga rata-rata lama menginapnya lebih pendek.
